```python
url = "https://raw.githubusercontent.com/eduardmendoza92/datawarehouse-seguros/main/data/clientes.csv"
```

```python
print("Registros duplicados")
print(df.duplicated().sum())
```

```python
print("Duplicados por nombre, email, fecha_nacimiento")
print(df.duplicated(subset=["nombre", "email", "fecha_nacimiento"]).sum())
```

```python
#Elimina espacios al inicio y al final en las columnas de texto usando for en una sola linea

for col in df.select_dtypes(include=["object"]):
    df[col] = df[col].str.strip()
```

```python
#Normalizar el nombre
df["nombre"] = df["nombre"].str.title()
```

```python
#Normalizar en minuscula el email
df["email"] = df["email"].str.lower()
```

```python
#Normalizar en mayuscula el genero
df["genero"] = df["genero"].str.upper()
```

```python
#Convertir fecha de nacimiento al tipo fecha
df["fecha_nacimiento"] = pd.to_datetime(df["fecha_nacimiento"], errors="coerce")
```

```python
#Normalizar ciudad a tipo titulo
df["ciudad"] = df["ciudad"].str.title()
```

```python
#Campos segmento astype, strip, upper
df["segmento"] = df["segmento"].astype(str).str.strip().str.upper()
```
